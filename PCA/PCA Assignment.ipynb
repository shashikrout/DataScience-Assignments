```python
# Normalizing the  numerical data
wine_norm=scale(wine_ary)
wine_norm
```

##### PCA Implementation

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
# PCA Components matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
cumvar=np.cumsum(np.round(var,4)*100)
cumvar
```

```python
# Variance plot for PCA components 
plt.figure(figsize=(5, 5))
plt.plot(cumvar, color='blue')
plt.show()
```

```python
wineds_f=pd.concat([df['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
wineds_f
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=wineds_f)
```

#### Hierarchical Clustering

```python
#create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wineclust=df.copy()
wineclust['clustersid']=hclusters.labels_
wineclust
```

##### K-Means 

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# Plot K values range vs WCSS to get Elbow graph 
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm using K=3
clust3=KMeans(3,random_state=30).fit(wine_norm)
clust3
```

```python
clust3.labels_
```

```python
# Assign clusters to the data set
wine_ds=df.copy()
wine_ds['clust3id']=clust3.labels_
wine_ds
```

```python
wine_ds['clust3id'].value_counts()
```
